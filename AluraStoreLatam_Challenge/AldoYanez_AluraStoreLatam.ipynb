## a) **Importación de datos**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## b) **ANÁLISIS DE DATOS**

###1. Análisis de facturación



-  En este primer análisis, debes calcular el **ingreso total** de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
tiendas = {
    "Tienda 1" : tienda,
    "Tienda 2" : tienda2,
    "Tienda 3" : tienda3,
    "Tienda 4" : tienda4
}
```

```python
ingresos = {
    nombre: df["Precio"].sum()
    for nombre, df in tiendas.items()
    }

df_ingresos = pd.DataFrame.from_dict(
    ingresos,
    orient="index",
    columns=["Ingresos totales"]
)

total_ingresos = df_ingresos["Ingresos totales"].sum()

df_ingresos["% del total"] = (
    df_ingresos["Ingresos totales"] / total_ingresos * 100
)

df_ingresos = df_ingresos.sort_values("Ingresos totales", ascending=False)

print(f"El total de los ingresos es de {total_ingresos}\n")
df_ingresos
```

### 2. Ventas por categoría

- En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
resultados = []

for nombre, df in tiendas.items():
  conteo = df['Categoría del Producto'].value_counts()

  for Categoria, Cantidad in conteo.items():
    resultados.append({
        "Tienda": nombre,
        "Categoría": Categoria,
        "Cantidad vendida": Cantidad
    })

df_categorias = pd.DataFrame(resultados)
df_categorias

```

### 3. Calificación promedio de la tienda


- En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
promedios = {}

for nombre, df in tiendas.items():
  promedios[nombre] = df['Calificación'].mean()

pd.DataFrame.from_dict(promedios, orient="index", columns=["Calificación promedio"])
```

### 4. Productos más y menos vendidos

- En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
for nombre, df in tiendas.items():
  conteo = df['Producto'].value_counts().reset_index()
  conteo.columns = ['Producto', 'Cantidad']

  top_3 = conteo.iloc[0:3]
  bottom_3 = conteo.iloc[-3:]

  print(f"\n{nombre} - Más vendidos:")
  display(top_3)

  print(f"\n{nombre} - Menos vendidos:")
  display(bottom_3)

```

### 5. Envío promedio por tienda

Descripción:

- En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
for nombre, df in tiendas.items():
  promedio_envio = df['Costo de envío'].mean()
  print(f"{nombre}: Costo de envío promedio = {promedio_envio:.2f}")
```

### 1. Ingresos por tienda

```python
ingresos = {
    nombre: df["Precio"].sum()
    for nombre, df in tiendas.items()
}

df_ingresos = pd.DataFrame.from_dict(
    ingresos, orient="index", columns=["Ingresos"]
)

df_ingresos.plot(
    kind="bar",
    figsize=(8,5),
    title="Ingresos totales por tienda",
    xlabel="Tienda",
    ylabel="Ingresos",
    legend=False
    )
plt.show()
```

- Porcentaje de ingresos por tienda:

```python
df_ingresos.plot(
    kind="pie",
    y="Ingresos",
    figsize=(6,6),
    autopct="%1.1f%%",
    title="Participación porcentual de ingresos por tienda",
    startangle=90,
    legend=False
)
plt.axis("equal")
plt.show()
```

### 2. Categorías más vendidas

```python
for nombre, df in tiendas.items():
  df["Categoría del Producto"]\
    .value_counts()\
    .sort_values()\
    .plot(
      kind="barh",
      title=f"Categorías más vendidas - {nombre}",
      figsize=(8, 5),
      xlabel="Cantidad de ventas",
      ylabel="Categoría"
  )

  plt.show()
```

### 3. Calificaciones por tienda

```python
calificaciones = {
    nombre: df["Calificación"].mean()
    for nombre, df in tiendas.items()
}

df_calificaciones = pd.DataFrame.from_dict(
    calificaciones,
    orient="index",
    columns=["Calificación promedio"]
)

df_calificaciones.plot(
    kind="line",
    marker="o",
    title="Calificación promedio",
    xlabel="Tienda",
    ylabel="Calificación",
)

plt.show()
```

### **Extra: Análisis de desempeño demográfico.**

- Importando librería:

```python
import folium
from folium.plugins import HeatMap
```

- Creando Mapa

```python
df_total = pd.concat(tiendas.values(), ignore_index=True)

mapa = folium.Map(
    location=[
        df_total['lat'].mean(),
        df_total['lon'].mean()
    ],
    zoom_start=5
)

colores = {
    "Tienda 1": "red",
    "Tienda 2": "blue",
    "Tienda 3": "green",
    "Tienda 4": "purple"
}

for nombre, df in tiendas.items():

  heat_data = df[['lat', 'lon']].values.tolist()

  HeatMap(
      heat_data,
      name=f"Concentración ventas {nombre}",
      radius=20,
      blur=15,
      min_opacity=0.3
  ).add_to(mapa)

folium.LayerControl(collapsed=False).add_to(mapa)
mapa
```

### Análisis de muestreo:

Al analizar el mapa, hemos encontrado que las zonas de ventas entre todas las tiendas son las mismas, esto apunta a que, efectivamente, en caso de liquidar la Tienda 4 para cambiar el giro o directamente empezar otro emprendimiento no existiría un impacto tan agresivo sobre los ingresos de la empresa.
Este análisis también apunta a poner más énfasis en expandir el alcance de las ventas a otras zonas geográficas de manera de poder optimizar los procesos de negocio, afianzando la marca y acrecentando los ingresos.
